```python
import pandas as pd

# Load your dataset
df = pd.read_csv('C:/Users/jaag1/Documents/data set/new york data cleaning dataset/AB_NYC_2019.csv')

# Data Integrity: Consistency Checks and Accuracy
# Ensure consistent labels in 'room_type'
df['room_type'] = df['room_type'].replace({
    'Entire home/apt': 'Entire home/apt',
    'Private room': 'Private room',
    'Shared room': 'Shared room',
    'Hotel room': 'Hotel room'
})

# Missing Data Handling: Imputation and Decision Making
# Fill missing values in 'reviews_per_month' with the column mean
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].mean())

# Remove rows with missing critical information
df.dropna(subset=['host_id', 'price', 'room_type'], inplace=True)

# Duplicate Removal: Identify and Eliminate Duplicates
df.drop_duplicates(subset=['id'], keep='first', inplace=True)

# Standardization: Consistent Formatting
# Convert 'price' to float if necessary (assuming it was a string with currency symbols)
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)

# Outlier Detection: Identify and Address Outliers
# For simplicity, let's consider outliers in 'price'
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
outlier_lower_bound = Q1 - 1.5 * IQR
outlier_upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['price'] >= outlier_lower_bound) & (df['price'] <= outlier_upper_bound)]

# Save the cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)

print("Data cleaning complete!")

df.head(101)



```
