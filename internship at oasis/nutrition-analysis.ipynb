```python
import pandas as pd

df = pd.read_csv('C:/Users/jaag1/Documents/nutrition.csv')  

print(df.head())
print(df.info())
print(df.describe())

df = df.dropna()  
```

```python
# Calculate basic statistics for nutritional information
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode().iloc[0]
std_dev_values = df.std(numeric_only=True)

print(f'Mean:\n{mean_values}\n')
print(f'Median:\n{median_values}\n')
print(f'Mode:\n{mode_values}\n')
print(f'Standard Deviation:\n{std_dev_values}\n')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Split your data into three parts for each graph.

# First Graph: Bar chart for Calories
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Calories', data=df)
plt.title('Calories per Item')
plt.xticks(rotation=90)
plt.show()

```
